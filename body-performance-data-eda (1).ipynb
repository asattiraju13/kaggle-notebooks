```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as pex

%matplotlib inline
```

```python
df = pd.read_csv('/kaggle/input/body-performance-data/bodyPerformance.csv')
```

```python
df.head()
```

## Matplotlib and Seaborn Visualizations

```python
fig, axes = plt.subplots(ncols = 3, figsize=(10,4))
cats = ['body fat_%','diastolic','systolic']

for i in range(len(cats)):
    data = df.groupby('gender')[cats[i]].median()
    
    axes[i].bar('Male',data.loc['M'], color='#942de3')
    axes[i].bar('Female',data.loc['F'], color='#2f91ed')
    
    axes[i].set_ylabel('Median ' + cats[i])

fig.suptitle('Medians of Various Metrics Across Genders')
fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(ncols = 4, figsize=(15,6))
cats = ['gripForce','sit and bend forward_cm','sit-ups counts','broad jump_cm']
for i in range(len(cats)):
    sns.boxplot(x='gender',y=cats[i],data=df,palette='coolwarm',ax=axes[i],hue='class')
axes[0].set_ylim(0,75)
axes[1].set_ylim(-30,50)
fig.tight_layout()
plt.show()
```

```python
sns.scatterplot(x='height_cm',y='weight_kg',data=df,hue='class',palette='coolwarm')
plt.xlim(135,195)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.show()
```

## Plotly Visualizations

```python
pex.scatter_3d(data_frame=df,x='height_cm',y='weight_kg',z='body fat_%',color='class')
```

```python
# Explore 2D histograms of correlated features identified in the heatmap above.
for tup in [('broad jump_cm','gripForce'),('broad jump_cm','body fat_%'), ('sit-ups counts','body fat_%')]:
    fig = pex.density_heatmap(df,x=tup[0],y=tup[1],width=500, height=500)
    fig.show()
```

```python
pex.density_heatmap(df,x='sit-ups counts',y='broad jump_cm',facet_row='gender',facet_col='class',color_continuous_scale='viridis')
```

```python
fig = pex.scatter(df, x="diastolic", y="systolic", color="height_cm", size='weight_kg')
fig.show(width=300,height=100)
```
